```python
dataset["payment"].fillna(dataset["payment"].mode()[0], inplace=True)
dataset["pickup_zone"].fillna(dataset["pickup_zone"].mode()[0], inplace=True)
dataset["dropoff_zone"].fillna(dataset["dropoff_zone"].mode()[0], inplace=True)
dataset["pickup_borough"].fillna(dataset["pickup_zone"].mode()[0], inplace=True)
dataset["dropoff_borough"].fillna(dataset["dropoff_borough"].mode()[0], inplace=True)

```

```python
dataset.isnull().sum()
```

#It shows pickup_day, pickup_hour, pickup_month, pickup_year

```python
dataset["pickup"] = pd.to_datetime(dataset["pickup"])
dataset["pickup_day"] = dataset["pickup"].dt.day_name()  # Monday, Tuesday, etc.
dataset["pickup_hour"] = dataset["pickup"].dt.hour  # 0-23
dataset["pickup_month"] = dataset["pickup"].dt.month  # 1-12
dataset["pickup_year"] = dataset["pickup"].dt.year  
```

```python
dataset
```

```python
dataset[["pickup", "pickup_day", "pickup_hour", "pickup_month" ,"pickup_year"]].head()
```

```python
# Ensure the pickup and dropoff columns are in datetime format
dataset['pickup'] = pd.to_datetime(dataset['pickup'], errors='coerce')
dataset['dropoff'] = pd.to_datetime(dataset['dropoff'], errors='coerce')

# Create the trip_duration column in minutes
dataset['trip_duration'] = (dataset['dropoff'] - dataset['pickup']).dt.total_seconds() / 60.0

# Verify the new column
print(dataset[['pickup', 'dropoff', 'trip_duration']].head())
```

#Fill missing values again

```python

dataset['payment'].fillna('Unknown', inplace=True)
dataset['pickup_zone'].fillna('Unknown', inplace=True)
dataset['dropoff_zone'].fillna('Unknown', inplace=True)
dataset['pickup_borough'].fillna('Unknown', inplace=True)
dataset['dropoff_borough'].fillna('Unknown', inplace=True)
print(dataset.isnull().sum())


```

```python
# Summary statistics
summary_stats = dataset.describe()

# Check for outliers using IQR method
q1 = dataset['fare'].quantile(0.25)
q3 = dataset['fare'].quantile(0.75)
iqr = q3 - q1
outliers = dataset[(dataset['fare'] < (q1 - 1.5 * iqr)) | (dataset['fare'] > (q3 + 1.5 * iqr))]
outliers
```

#Peak Hours for Uber Rides

```python
rides_by_hour = dataset['pickup_hour'].value_counts().sort_index()
plt.figure()
sns.barplot(x=rides_by_hour.index, y=rides_by_hour.values, palette='viridis')
plt.title('Uber Rides by Pickup Hour')
plt.xlabel('Pickup Hour')
plt.ylabel('Number of Rides')
plt.show()

```

#Ride Demand Variation Across Boroughs (using pickup_borough)

```python
rides_by_borough = dataset['pickup_borough'].value_counts()
plt.figure()
sns.barplot(x=rides_by_borough.index, y=rides_by_borough.values, palette='plasma')
plt.title('Ride Demand by Pickup Borough')
plt.xlabel('Pickup Borough')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.show()


```

#Average Trip Duration by Pickup Zone

```python
avg_duration_by_zone = dataset.groupby('pickup_zone')['trip_duration'].mean().sort_values(ascending=False)
print("Average Trip Duration by Pickup Zone (in minutes):")
print(avg_duration_by_zone)
plt.figure()
avg_duration_by_zone.plot(kind='bar', color='teal')
plt.title('Average Trip Duration by Pickup Zone')
plt.xlabel('Pickup Zone')
plt.ylabel('Average Trip Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```

#correlation between fare, distance, trip duration, and tolls


```python
corr_matrix = dataset[['fare', 'distance', 'trip_duration', 'tolls']].corr()
print("Correlation Matrix:")
print(corr_matrix)
plt.figure()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Fare, Distance, Trip Duration, and Tolls')
plt.show()
```

#Tip Percentage

```python
dataset['tip_percentage'] = np.where(dataset['fare'] > 0, dataset['tip'] / dataset['fare'] * 100, 0)
plt.figure()
sns.histplot(dataset['tip_percentage'], bins=30, kde=True, color='purple')
plt.title('Tip Percentage Distribution')
plt.xlabel('Tip Percentage (%)')
plt.ylabel('Frequency')
plt.show()
```

#Impact of Tolls on Total Fare


```python
plt.figure()
sns.scatterplot(x='tolls', y='total', data=dataset, alpha=0.6, color='green')
plt.title('Impact of Tolls on Total Fare')
plt.xlabel('Tolls ($)')
plt.ylabel('Total Fare ($)')
plt.show()
```

#Most Commonly Used Payment Method


```python
payment_counts = dataset['payment'].value_counts()
print("Payment Method Distribution:")
print(payment_counts)
plt.figure()
sns.barplot(x=payment_counts.index, y=payment_counts.values, palette='pastel')
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

#Comparing Ride Demand by Borough


```python
pickup_borough_counts = dataset['pickup_borough'].value_counts()
dropoff_borough_counts = dataset['dropoff_borough'].value_counts()

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.barplot(x=pickup_borough_counts.index, y=pickup_borough_counts.values, ax=ax[0], palette='coolwarm')
ax[0].set_title('Pickup Borough Ride Demand')
ax[0].set_xlabel('Borough')
ax[0].set_ylabel('Number of Rides')
ax[0].tick_params(axis='x', rotation=45)

sns.barplot(x=dropoff_borough_counts.index, y=dropoff_borough_counts.values, ax=ax[1], palette='coolwarm')
ax[1].set_title('Drop-off Borough Ride Demand')
ax[1].set_xlabel('Borough')
ax[1].set_ylabel('Number of Rides')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
